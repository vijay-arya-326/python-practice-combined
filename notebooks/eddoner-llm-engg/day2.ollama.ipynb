```python
import random
from dotenv import load_dotenv
from pathlib import Path
import sys
import os

cwd = Path.cwd().parent.parent.joinpath("env/.env")

if cwd.exists():
    load_dotenv(dotenv_path=cwd, override=True)
else:
    print("No .env file found")
    sys.exit(1)

print(os.environ["APP_NAME"])
```

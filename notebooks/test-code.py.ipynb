```python
import functools

def print_io(func):
    @functools.wraps(func)
    def inner(*args, **kwargs):
        print("Input args:", args, "kwargs:", kwargs)
        result = func(*args, **kwargs)
        print("Output:", result)
        return result
    return inner

@print_io
def add(a, b):
    return a + b

@print_io
def greet(name="World"):
    return f"Hello, {name}!"

add(2, 3)
greet(name="Alice")
```

```python
def combine(a, b):
    if isinstance(a, (int, float)) and isinstance(b, (int, float)):
        return a + b
    if isinstance(a, str) or isinstance(b, str):
        return str(a) + str(b)
    raise TypeError("Both values must be numbers or strings")

# examples
combine(2, "Ascacsa")
```

```python
import datetime
import os
import jwt
import copy
import sys

# In a real application, this secret key should be stored securely (e.g., environment variable)
# and be much more complex.
SECRET_KEY = os.getenv("JWT_SECRET_KEY", None)

if not SECRET_KEY:
    print("No JWT secret key found")
    sys.exit(0)

ALGORITHM = "HS256"

def create_jwt_token(data: dict):
    to_encode = data.copy()
    to_encode_without_expiry = copy.deepcopy(to_encode)
    # Add expiration time
    expire = datetime.datetime.now(datetime.timezone.utc) + datetime.timedelta(minutes=30)
    to_encode.update({"exp": expire})
    encoded_jwt = jwt.encode(to_encode, SECRET_KEY, algorithm=ALGORITHM)
    encoded_access_token = jwt.encode(to_encode_without_expiry, SECRET_KEY, algorithm=ALGORITHM)
    return [encoded_jwt, encoded_access_token]

def login(username, password):
    # In a real application, you would verify the username and password against a database
    if username == "user" and password == "password":
        # If authentication is successful, create a JWT token
        token_data = {"sub": username, "role": "user"}
        [access_token, refresh_token] = create_jwt_token(token_data)
        return {"message": "Login successful", "access_token": access_token, "refresh_token": refresh_token, "token_type": "bearer"}
    else:
        return {"message": "Invalid credentials"}

def jwt_decode(data:str):
    return jwt.decode(data, SECRET_KEY, algorithms=[ALGORITHM])

# Example usage:
result =  login("user", "password")
print(jwt_decode(result["access_token"]))
print(jwt_decode(result["refresh_token"]))
# print(login("wrong_user", "wrong_password"))
```
